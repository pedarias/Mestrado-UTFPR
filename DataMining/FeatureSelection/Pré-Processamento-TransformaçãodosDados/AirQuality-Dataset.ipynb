### Tratamento de Data e Hora
* Base de Dados: Air Quality Dataset
* https://www.kaggle.com/datasets/citrahsagala/airquality

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('AirQualityUCI.csv')
df
```

```python
df.columns
```

```python
df.info()
```

```python
# Conversão de object para  datetime
df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
```

```python
df['Date']
```

```python
# Criando novas colunas com informações de ano, mês e dia.
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.day_of_week
```

```python
df.info()
```

```python
# Conversão de hora (object) para datetime
df['Time'] = pd.to_datetime(df['Time'], format = '%H.%M.%S')
```

```python
# Obtendo apenas a hora
df['Hour'] = df['Time'].dt.hour
```

```python
# Remoção da coluna Time
df.drop(['Time'], axis=1, inplace=True)
```

```python
df.info()
```

#### Análise Exploratória

```python
# Breve Análise Exploratória dos dados (Box Plot)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 3))

def create_boxplot(feature):
    sns.boxplot(data = df[[feature]])
    plt.title(feature, fontsize = 10)

plt.subplot(1,3,1)
create_boxplot('CO(GT)')

plt.subplot(1,3,2)
create_boxplot('T')

plt.subplot(1,3,3)
create_boxplot('RH')

plt.show()
```

```python
# Verifica a quantidade de valores iguaç -200
df['T'].loc[df['T'] <= -100].size
```

```python
#  Calcula a média das temperaturas (sem os valores -200)
mediaT = df['T'].loc[df['T']>=0].mean()
mediaCO = df['CO(GT)'].loc[df['CO(GT)']>=0].mean()
mediaRH = df['RH'].loc[df['RH']>=0].mean()
```

```python
# Substituir valores dos atributos pela média
# Linhas/Coluna
df.loc[df['T']<0, 'T'] = mediaT
df.loc[df['CO(GT)']<0, 'CO(GT)'] = mediaCO
df.loc[df['RH']<0, 'RH'] = mediaRH
```

```python
plt.figure(figsize=(10, 3))

plt.subplot(1,3,1)
create_boxplot('CO(GT)')

plt.subplot(1,3,2)
create_boxplot('T')

plt.subplot(1,3,3)
create_boxplot('RH')

plt.show()
```
