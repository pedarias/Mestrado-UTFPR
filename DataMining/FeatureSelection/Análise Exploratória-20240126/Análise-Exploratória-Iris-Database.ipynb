# Visualização dos Dados

Remoção do atributo ID para não influenciar na análise exploratória dos dados

```python
df.drop(columns = ['Id'], axis = 1, inplace = True)
df
```

#### (1) `Box Plot`
* Resumo das propriedades de um conjunto de valores;
* Ex: mínimo, primeiro quartil, mediana, terceiro quartil e máximo

```python
plt.figure(figsize=(10, 10))

def create_boxplot(feature):
    sns.boxplot(data = df, x = 'Species', y = feature)
    plt.title(feature, fontsize = 10)
    plt.xlabel('Species', fontsize = 10)
    plt.ylabel(feature, fontsize = 10)

plt.subplot(2,2,1)
create_boxplot('SepalLengthCm')

plt.subplot(2,2,2)
create_boxplot('SepalWidthCm')

plt.subplot(2,2,3)
create_boxplot('PetalLengthCm')

plt.subplot(2,2,4)
create_boxplot('PetalWidthCm')

plt.show()
```

#### (2) `Violin Plot`
-  ViolinPlot combina a visualização do boxplot juntamente com um plot de densidade (similar ao que vemos nos histogramas);
- Dentro do ViolinPlot encontramos os mesmos elementos que encontramos no BoxPlot;
- Mais detalhes: https://towardsdatascience.com/violin-plots-explained-fb1d115e023d
<span style="width:80%;text-align: center;"><img align = left src="https://miro.medium.com/v2/resize:fit:1040/format:webp/1*TTMOaNG1o4PgQd-e8LurMg.png" style="height:550px;"></span>

```python
plt.figure(figsize=(10, 10))

def create_violinplot(feature):
    sns.violinplot(data = df, x = 'Species', y = feature)
    plt.title(feature, fontsize = 10)
    plt.xlabel('Species', fontsize = 10)
    plt.ylabel(feature, fontsize = 10)

plt.subplot(2,2,1)
create_violinplot('SepalLengthCm')

plt.subplot(2,2,2)
create_violinplot('SepalWidthCm')

plt.subplot(2,2,3)
create_violinplot('PetalLengthCm')

plt.subplot(2,2,4)
create_violinplot('PetalWidthCm')

plt.show()
```

#### (3) `Scatterplot`
- Análise entre dois atributos


```python
plt.figure(figsize = (5, 3))
sns.scatterplot(x = 'PetalLengthCm', y = 'PetalWidthCm', hue = 'Species', data = df, s = 65)
plt.title('Petal Length x Petal Width', fontsize = 15)
plt.xlabel('Petal Length', fontsize = 15)
plt.ylabel('Petal Width', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(x = 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', data = df, s = 65)
plt.title('Sepal Length x Sepal Width', fontsize = 15)
plt.xlabel('Sepal Length', fontsize = 15)
plt.ylabel('Sepal Width', fontsize = 15)
plt.show()
```

#### (4) `pairplot`
* Análise de todos atributos


```python
# sns.pairplot(df, hue = 'Species', diag_kind = 'hist', height=5)
sns.pairplot(df, hue = 'Species', diag_kind = 'hist', height=2)
plt.show()
```

#### (5) `Jointplot`
- O jointplot é a junção de um gráfico univariado, nesse caso um histograma, e um bivariado, um scatterplot.
- Análise entre os atributos **SepalLengthCm** e **SepalWidthCm**

```python
sns.jointplot(x = 'SepalLengthCm', y = 'SepalWidthCm', data = df, hue = 'Species', palette = 'PuRd')
```

#### (6) `Heatmap`
* Avaliando correlações


```python
plt.figure(figsize = (15, 10))
sns.heatmap(df.corr(), cmap = 'Blues', square = True, annot = True)
plt.title("Visualizing Correlations", size = 20)
plt.show()
```

#### (7) `Histplot`
* Frequência dos dados

```python
plt.figure(figsize=(10, 10))

def create_histogram(feature):
    sns.histplot(data = df, hue = 'Species', x = feature, kde=False)

    plt.title('Boxplot for ' + feature, fontsize = 10)
    plt.xlabel(feature, fontsize = 10)
    plt.ylabel('Frequência', fontsize = 10)

plt.subplot(2,2,1)
create_histogram('SepalLengthCm')

plt.subplot(2,2,2)
create_histogram('SepalWidthCm')

plt.subplot(2,2,3)
create_histogram('PetalLengthCm')

plt.subplot(2,2,4)
create_histogram('PetalWidthCm')

plt.show()
```
