# Visualização dos Dados

```python
df.columns
```

#### (1) `Box Plot`
* Resumo das propriedades de um conjunto de valores;
* Ex: mínimo, primeiro quartil, mediana, terceiro quartil e máximo

```python
plt.figure(figsize=(10, 14))

def create_boxplot(feature):
    sns.boxplot(data = df, x = 'class', y = feature)
    plt.xlabel('class', fontsize = 10)
    plt.ylabel(feature, fontsize = 10)

plt.subplot(3,2,1)
create_boxplot('age')

plt.subplot(3,2,2)
create_boxplot('impluse')

plt.subplot(3,2,3)
create_boxplot('pressurehight')

plt.subplot(3,2,4)
create_boxplot('glucose')

plt.subplot(3,2,5)
create_boxplot('kcm')

plt.subplot(3,2,6)
create_boxplot('troponin')



plt.show()
```

#### (2) `Violin Plot`
-  ViolinPlot combina a visualização do boxplot juntamente com um plot de densidade (similar ao que vemos nos histogramas);
- Dentro do ViolinPlot encontramos os mesmos elementos que encontramos no BoxPlot;
- Mais detalhes: https://towardsdatascience.com/violin-plots-explained-fb1d115e023d
<span style="width:80%;text-align: center;"><img align = left src="https://miro.medium.com/v2/resize:fit:1040/format:webp/1*TTMOaNG1o4PgQd-e8LurMg.png" style="height:550px;"></span>

```python
plt.figure(figsize=(10, 14))

def create_violinplot(feature):
    sns.violinplot(data = df, x = 'class', y = feature)
    plt.xlabel('class', fontsize = 10)
    plt.ylabel(feature, fontsize = 10)

plt.subplot(3,2,1)
create_violinplot('age')

plt.subplot(3,2,2)
create_violinplot('impluse')

plt.subplot(3,2,3)
create_violinplot('pressurehight')

plt.subplot(3,2,4)
create_violinplot('glucose')

plt.subplot(3,2,5)
create_violinplot('kcm')

plt.subplot(3,2,6)
create_violinplot('troponin')
plt.show()
```

#### (3) `Scatterplot`
- Análise entre dois atributos


```python
df.columns
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(x = 'pressurehight', y = 'glucose', hue = 'class', data = df, s = 65)
plt.title('pressurehight x glucose', fontsize = 15)
plt.xlabel('pressurehight', fontsize = 15)
plt.ylabel('glucose', fontsize = 15)
plt.show()
```

#### (4) `pairplot`
* Análise de todos atributos


```python
# sns.pairplot(df, hue = 'Species', height = 5)
sns.pairplot(df, hue = 'class', diag_kind = 'hist', height=5)
plt.show()
```

#### (5) `Jointplot`
- O jointplot é a junção de um gráfico univariado, nesse caso um histograma, e um bivariado, um scatterplot.

```python
sns.jointplot(x = 'pressurehight', y = 'glucose', data = df, hue = 'class', palette = 'PuRd')
```

#### (6) `Heatmap`
* Avaliando correlações


```python
plt.figure(figsize = (15, 10))
sns.heatmap(df.corr(), cmap = 'Blues', square = True, annot = True)
plt.title("Visualizing Correlations", size = 20)
plt.show()
```

#### (7) `Histplot`
* Frequência dos dados

```python
plt.figure(figsize=(10, 14))

def create_histogram(feature):
    sns.histplot(data = df, x = feature, kde=True)
    plt.xlabel(feature, fontsize = 10)
    plt.ylabel('Frequência', fontsize = 10)

plt.subplot(3,2,1)
create_histogram('age')

plt.subplot(3,2,2)
create_histogram('impluse')

plt.subplot(3,2,3)
create_histogram('pressurehight')

plt.subplot(3,2,4)
create_histogram('glucose')

plt.subplot(3,2,5)
create_histogram('kcm')

plt.subplot(3,2,6)
create_histogram('troponin')
plt.show()

```
