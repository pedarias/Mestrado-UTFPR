##  Tratamento dos dados

### Feature Engineering

```python
df = df.drop('Booking_ID', axis =1)
df = df.drop('arrival_year', axis = 1)
```

Vou dropar o Booking ID e o ano da reserva para evitar overfitting, já que o ID poderia acabar indicando uma relação entre as reservas e o ano é uma feature que não vai se repetir nos anos seguintes, não sendo útil para a identificação de cancelamentos futuros

```python
df['type_of_meal_plan'] = df['type_of_meal_plan'].replace({'Not Selected': 0, 'Meal Plan 1': 1, 
                                                           'Meal Plan 2': 2, 'Meal Plan 3': 3})
df['room_type_reserved'] = df['room_type_reserved'].replace({
    'Room_Type 1': 1,
    'Room_Type 2': 2,
    'Room_Type 3': 3,
    'Room_Type 4': 4,
    'Room_Type 5': 5,
    'Room_Type 6': 6,
    'Room_Type 7': 7
})
df['market_segment_type'] = df['market_segment_type'].replace({'Offline': 1, 'Online': 2, 
                                                               'Aviation': 3, 'Corporate': 4, 
                                                               'Complementary': 5})
```

É preciso converter as strings em valores numéricos para poder funcionar dentro dos modelos de machine learning

```python
df = df.copy()  # copiar a base de dados df
df['booking_status'] = pd.get_dummies(df['booking_status'])['Canceled']
df.head()
```

Transformendo a variável alvo em dummie (0-1)

### Split treino e teste

```python
y = df["booking_status"]
X = df.drop("booking_status", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

## Modelos

Os algorítimos usados serão:
- **Regressão Logística:** Utiliza máxima verossimilhança ára estimar uma função sigmoide que irá classificar a variável alvo. No geral é um algorítimo simples mas utilizado em alvos binários
- **K-NEAREST NEIGHBORS(KNN):** Funciona selecionando a classe majoritária dos K-"vizinhos" mais próximos apartir da distância (euclidiana, Ponderada, Manhattan ou Minkowski). Simples e efetivo mas acaba sendo muito sensível a pequenas variações
- **Naive Bayes((NB):** Utiliza o Teorema Baisiano par gerar previsões nos dados a partir da incorporação de eventos anteriores seguindo a Lei de Bayes. Tem como maior problema a premissa d que as variáveis são independentes
- **Árvore de Decisão:** A partir de nodes (espécies de nós com perguntas sobre os dados) esse algorítimo vai classificar a partir da respostas que for obtendo para cada nó
- **Random Forest:** Une várias árvores de decisão através de agregação de boostraps e depois realiza um splot usando menos variáveis. É vantajoso porque não tem alta sensibilidade a novos dados e consegue manter bem a acurácia 

```python
models = {'Logistic Regression': LogisticRegression(),
          'KNN': KNeighborsClassifier(),
          'Naive Bayes': GaussianNB(),
          'Decision Tree': DecisionTreeClassifier(),
          'Random Forest': RandomForestClassifier()}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

for name, acc in results.items():
    print(f"{name}: {acc:.3f}")
```

### Comparando os modelos

```python
fig, ax = plt.subplots(figsize=(10,6))

x = list(results.keys())
y = list(results.values())

ax.bar(x, y)

ax.set_title('Comparação de acurácias entre modelos')
ax.set_xlabel('Modelo')
ax.set_ylabel('Acurácia')

plt.show()
```

Como a métrica de sucesso escolhida foi a Acurácia, o modelo que obteve a melhor performance foi o Random Forest
