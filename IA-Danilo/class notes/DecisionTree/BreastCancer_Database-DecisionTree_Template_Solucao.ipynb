#### Pré-processamento

```python
from sklearn.preprocessing import LabelEncoder
```

```python
len(dados['id'].unique())
```

```python
dados['Unnamed: 32']
```

```python
# Remove atributos irrelevantes da base
# Método drop (pandas)
dados.drop(['Unnamed: 32', 'id'], axis=1, inplace=True)
```

```python
dados.columns
```

```python
dados.info()
```

```python
le = LabelEncoder()
dados['diagnosis'] = le.fit_transform(dados['diagnosis'])
```

```python
dados['diagnosis'].unique()
```

```python
dados.info()
```

#### Separação entre treino e teste

```python
from sklearn.model_selection import train_test_split
```

```python
X = dados.drop(['diagnosis'], axis=1).values
```

```python
y = dados['diagnosis'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7, test_size=0.3)
```

#### Algoritmo Árvore de Decisão

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(criterion="entropy")
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

### Importando métricas do Sklearn

```python
from sklearn.metrics import accuracy_score
```

### Acuracidade

```python
accuracy_score(y_test, y_pred)
```

### Matriz Confusão

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
```

#### Plotando a árvore

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
fig = plt.figure(figsize=(20,8))

tree.plot_tree(clf,
                   feature_names=dados.columns.drop('diagnosis'),
                   class_names=['benign', 'malignant'],
                   filled=False)

# tree.plot_tree(clf)

plt.show()
```
